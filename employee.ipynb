```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dept_df = pd.read_csv("data/departments.csv", index_col=False)
dept_df.reset_index(drop=True, inplace=True)
dept_df
```

```python
emp_df = pd.read_csv("data/dept_emp.csv", index_col=False)
emp_df.reset_index(drop=True, inplace=True)
emp_df
```

```python
mgr_df = pd.read_csv("data/dept_manager.csv", index_col=False)
mgr_df.reset_index(drop=True, inplace=True)
mgr_df
```

```python
employees_df = pd.read_csv("data/employees.csv", index_col=False)
employees_df.reset_index(drop=True, inplace=True)
employees_df
```

```python
salaries_df = pd.read_csv("data/salaries.csv", index_col=False)
salaries_df.reset_index(drop=True, inplace=True)
salaries_df
```

```python
titles_df = pd.read_csv("data/titles.csv", index_col=False)
titles_df.reset_index(drop=True, inplace=True)
titles_df
```

```python
new_df = pd.merge(employees_df, salaries_df, how='left', on=['emp_no', 'emp_no'])
```
